```python
with open('../input/2022-cricket-world-cup/t20_wc_match_results.json') as f:
    data = json.load(f)
#data(len) #output is 1 that confirms only one element
#data[0]['matchSummary'] #give list of match result
# data array has only one element i.e match summary 
df_match = pd.DataFrame(data[0]['matchSummary'])  #creates a DataFrame named df_match from the "matchSummary" key of the first element in the data list
df_match.head()
```

```python
df_match.shape  #shows shape 45 rows and 7 column
```

```python
#We want to use scorecard column as the primary key when working in powerbi that,s why we will rename the column name to match id
df_match.rename({'scorecard':'match_id'},axis=1,inplace = True)
#The axis=1 argument indicates that the renaming should be done along the columns (i.e., renaming column labels)
#the inplace=True argument ensures that the changes are made directly to the DataFrame df_match without creating a new DataFrame.
df_match.head()
```

#As batting summary table and match result table has only team 1 and team 2 name common so that can act as primry key for both tables. but Match result table has team name in two diff columns while batting summary has team names in one column in format 'team1 vs team 2'.
so to make primary key we need to join team1 and team 2 column of match result table with vs in between.
the match id will be same for team1 vs team2 and team2 vs team1. so we are creating dictionary of teams and match id

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

Batting Summary**

```python
 with open('../input/2022-cricket-world-cup/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    all_records = []
# all the matches is presented by one single json element called battingSummary inside array    
    for rec in data: 
        all_records.extend(rec['battingSummary'])
#each record is one match and one match has multiple record for the player
#rec['battingSummary'] - it will give batting summary of one match 
#to join all records in a single flat list we have created a list called all records and then extend function is used to join all record array 
        
df_batting = pd.DataFrame(all_records)
df_batting.head()
```

```python
df_batting["out/not_out"] =df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
#apply() method in combination with a lambda function is used to create a new column called "out/not_out" in the DataFrame df_batting
#"out" if len(x)>0 else "not_out" - because not out column is blank and out has some dismissal value
df_batting.head()
```

```python
df_batting.drop(columns=["dismissal"],inplace=True) 
# we r dropping dismissal column because we have added out notout column
df_batting.head()
```

```python
#to cleanup the weired characters like â€ or \xa0 in the batsman name 
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head()
```

```python
# match_ids_dict["Namibia Vs Sri Lanka"]
# output - 'T20I # 1823'
```

we will be creating new column match id to batting summary table

```python
df_batting["match_id"] =df_batting["match"].map(match_ids_dict)
df_batting.head()

#The map() method is used to apply the match_ids_dict to each value in the "match" column. It looks up each value in the dictionary and replaces it with the corresponding match ID
```

```python
#creating folder in the kaggle environment to save file
if not os.path.exists('t20_csv_files'):
    os.makedirs('t20_csv_files')
```

```python
#Export the table to csv format in folder t20_csv_files 
df_batting.to_csv('t20_csv_files/fact_bating_summary.csv', index = False)

#The index=False argument specifies that the DataFrame index should not be included in the CSV file.
```

```python
with open('../input/2022-cricket-world-cup/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```
