```python
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

```python


def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol) if mol != None else print(elem,mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:
        #print(mol)        
       
        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)
           
        row = np.array([desc_MolWt,
                        desc_MolLogP,
                        desc_NumHDonors,
                        desc_NumHAcceptors])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
```

```python
df_lipinski = lipinski(new_df.canonical_smiles,verbose=False)
df_lipinski.tail()
```

```python
import numpy as np
import math

def pIC50(input):
    pIC50 = []

    for i in input['standard_value']:
      pic50 =  6 - math.log10(i)
      pIC50.append(pic50)

        
    return pd.DataFrame(pIC50)
```

# **Xgboost regression**

```python
import pandas as pd
import numpy as np

df = pd.read_csv('cancer_pIC50_pubchem_fp.csv')
X = df.drop('pIC50', axis=1)
Y = df.pIC50
Y = Y.dropna()

# Removing the low variance data
from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(threshold=(.8 * (1 - .8)))    
X = selection.fit_transform(X)

#Split the data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)


###### Xgboost Regression in Python #######
from xgboost import XGBRegressor
RegModel=XGBRegressor(max_depth=8, learning_rate=0.1, n_estimators=1000, booster='gbtree')

#Printing all the parameters of XGBoost
print(RegModel)

#Creating the model on Training Data
XGB=RegModel.fit(X_train,Y_train)
prediction=XGB.predict(X_test)

#Measuring Goodness of fit in Training data
from sklearn import metrics
print('R2 Value : ',metrics.r2_score(Y_train, XGB.predict(X_train)))  #it predict coorelation between x and y variable,more corelation more goodness of fit
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_test, prediction)
print("MSE: ", mse)

#Measuring accuracy on Testing Data
print('Accuracy : ',100- (np.mean(np.abs((Y_test - prediction) / Y_test)) * 100))

```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_test, prediction)
rmse = np.sqrt(mse)
print("RMSE : ",rmse)

from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(Y_test, prediction)
print("MAE : ",mae)
```

# **AdaBoost regression**

```python
import pandas as pd
import numpy as np

df = pd.read_csv('cancer_pIC50_pubchem_fp.csv')
X = df.drop('pIC50', axis=1)
Y = df.pIC50
Y = Y.dropna()



#Split the data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

###### Adaboost Regression in Python #######
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor

#Choosing Decision Tree with 1 level as the weak learner
DTR=DecisionTreeRegressor(max_depth=7)
RegModel = AdaBoostRegressor(n_estimators=1000, base_estimator=DTR ,learning_rate=0.1)

#Printing all the parameters of Adaboost
print(RegModel)

#Creating the model on Training Data
AB=RegModel.fit(X_train,Y_train)
prediction=AB.predict(X_test)

#Measuring Goodness of fit in Training data
from sklearn import metrics
print('R2 Value:',metrics.r2_score(Y_train, AB.predict(X_train)))

#Measuring accuracy on Testing Data
print('Accuracy',100- (np.mean(np.abs((Y_test - prediction) / Y_test)) * 100))

```
